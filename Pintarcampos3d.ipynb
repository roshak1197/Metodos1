```python
%matplotlib notebook
import numpy as np 
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import axes3d
```

```python
#vamos a definir un movimiento circular y un radio 
A  = 1
B = 3
omega = 2*np.pi/10
N = 100
t = np.linspace(0,10,N)
r = np.zeros((N,3)) #define la posicion en 3d es una matriz!
r #podemos print esto solo para ver qué estamos haciendo
```

```python
r[:,0] = A*np.cos(omega*t) #esta es la mata del for loop
r[:,1] = A*np.sin(omega*t)
r[:,2] = B*t
```

```python
fig = plt.figure(figsize=(5,5))
ax = fig.add_subplot(111,projection='3d')
ax.scatter(r[:,0],r[:,1],r[:,2],color='k',marker='o',s=100)
ax.view_init()
```

```python
#pintando un campo vectorial 
n = 4
x = np.linspace(0,1,n)
y = np.linspace(0,1,n)
X,Y = np.meshgrid(x,y) #organiza d manera matricial las coordenadas de x e y
X
```

```python
r0 = np.array([0.4,0.5])
def GetE(r,r0,q=1,c=0):
    d = (r[0]- r0[0])**2 + (r[1]-r0[1])**2
    return q*r[c]/d**(3/2)
```

```python
Ex = np.zeros((n,n))
Ey = np.zeros((n,n))#llenamos el campo
for i in range(n):
    for j in range(n):
        Ex[i,j] = GetE.
        Ey[i,j] = -2.
```

```python
fig = plt.figure(figsize=(5,5)) #dispersion del campo partido en 4 cada coordenada
ax = fig.add_subplot(111)
ax.scatter(X,Y)
for i in range(n):
    for j in range(n):
        ax.quiver(x[i],y[j],Ex[i,j],Ey[i,j])
```
